```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
kospi = pd.read_csv('./data/kospi.csv', encoding='cp949')
```

```python
kospi.head()
```

```python
kospi.info()
```

```python
y = kospi['코스피지수']
X = kospi.drop('코스피지수', axis=1)
```

```python
pd.options.display.max_rows = 200
pd.options.display.max_columns = 200
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# 1) DecisionTreeRegressor
from sklearn.tree import DecisionTreeRegressor
dtr = DecisionTreeRegressor()
dtr.fit(X_train, y_train)
dtr_pred = dtr.predict(X_test)

from sklearn.metrics import mean_squared_error
dtr_rmse = mean_squared_error(y_test, dtr_pred, squared=False)
print(f'dtr rmse : {dtr_rmse}')
```

```python
for d in range(1,10):
    dtr = DecisionTreeRegressor(max_depth = d, random_state=1245)
    dtr.fit(X_train, y_train)
    dtr_pred = dtr.predict(X_test)
    dtr_rmse = mean_squared_error(y_test, dtr_pred, squared=False)
    print(f'max_depth = {d}, dtr rmse : {dtr_rmse}')
```

```python
# 2) RandomForestRegressor
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor(random_state=42)
rfr.fit(X_train, y_train)
rfr_pred = rfr.predict(X_test)
rfr_rmse = mean_squared_error(y_test, rfr_pred, squared=False)
rfr_rmse
```

```python
for d in range(1,10):
    rfr = RandomForestRegressor(max_depth = d, random_state=1004)
    rfr.fit(X_train, y_train)
    rfr_pred = rfr.predict(X_test)
    rfr_rmse = mean_squared_error(y_test, rfr_pred, squared=False)
    print(f'max_depth = {d}, rfr rmse : {rfr_rmse}')
```

```python
# 3) BaggingRegressor
from sklearn.ensemble import BaggingRegressor
bgr = BaggingRegressor(random_state=42)
bgr.fit(X_train, y_train)
bgr_pred = bgr.predict(X_test)
bgr_rmse = mean_squared_error(y_test, bgr_pred, squared=False)
bgr_rmse
```

```python
for est in range(10,100,10):
    bgr = BaggingRegressor(random_state=42, n_estimators=est)
    bgr.fit(X_train, y_train)
    bgr_pred = bgr.predict(X_test)
    bgr_rmse = mean_squared_error(y_test, bgr_pred, squared=False)
    print(f'n_estimator = {est}, bgr rmse : {bgr_rmse}')
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.3 ,random_state = 536)
best_model = RandomForestRegressor(n_estimators=100, max_depth = 11,random_state = 536)
best_model.fit(X_train, y_train)
pred = best_model.predict(X_test)
print(mean_squared_error(y_test, pred, squared=False))
```

```python
for d in range(1,15):
    rfr = RandomForestRegressor(max_depth = d, random_state=536)
    rfr.fit(X_train, y_train)
    rfr_pred = rfr.predict(X_test)
    rfr_rmse = mean_squared_error(y_test, rfr_pred, squared=False)
    print(f'max_depth = {d}, rfr rmse : {rfr_rmse}')
```

```python
X_train2, X_val, y_train2, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=42)
```

```python
from itertools import product
est = range(100,800, 100)    # 보통 100단위
depth = range(3,10,2)               # 보통 홀수 기준
grid = list(product(est, depth))
```

```python
best_param = None
best_rmse = 999999999  # 최대값을 찾기 위한 변수 초기화
for param in grid:      # param : tuple 타입
    print(f'{param} 서칭중')
    rfr = RandomForestRegressor(n_estimators=param[0], max_depth=param[1], random_state=1004)
    rfr.fit(X_train2, y_train2)
    rfr_pred = rfr.predict(X_val)
    score = mean_squared_error(y_val, rfr_pred, squared=False)
    if best_rmse > score:
        best_rmse = score
        best_param = param
    print()
```
